```python
import matplotlib.pyplot as plt
import numpy as np
mark1 = np.array([2, 4, 5, 7, 9, 10, 11, 13, 14, 15])
mark2 = np.array([6, 7, 8, 10, 11, 12, 13, 14, 15, 16])
plt.hist([mark1, mark2]), bins=10, stacked=True, color=['skyblue', 'salmon'], edgecolor='black')
plt.xlabel('marks')
plt.ylabel('count')
plt.title('stacked histogram with multiple datasets')
plt.legend('markdataset1')
plt.show()
```
